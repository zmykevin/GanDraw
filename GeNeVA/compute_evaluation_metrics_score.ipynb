```python
import os
import glob
import cv2
import matplotlib.pyplot as plt
```

```python
result_path = "logs/gandraw_baseline1/results"

total_scene = 90
for i in range(total_scene):
    pred_path = os.path.join(result_path, str(i))
    gt_path = os.path.join(result_path, "{}_gt".format(str(i)))
    
    #load the all the images
    pred_image_list = glob.glob(pred_path+"/*.png")
    gt_image_list = glob.glob(gt_path+"/*.png")
    if len(pred_image_list) > 0 and len(gt_image_list) > 0:
        print("predicted image:")
        pred_image = cv2.imread(pred_image_list[0])
        pred_image = cv2.cvtColor(pred_image, cv2.COLOR_BGR2RGB)
        plt.imshow(pred_image)
        plt.show()
        #print(gt_image_list)
        print("ground_truth image:")
        gt_image = cv2.imread(gt_image_list[0])
        gt_image = cv2.cvtColor(gt_image, cv2.COLOR_BGR2RGB)
        plt.imshow(gt_image)
        plt.show()
        
        gt_image_2 = cv2.imread(gt_image_list[1])
        gt_image_2 = cv2.cvtColor(gt_image_2, cv2.COLOR_BGR2RGB)
        plt.imshow(gt_image_2)
        plt.show()
    
    if i == 10:
        break
    
  
```
